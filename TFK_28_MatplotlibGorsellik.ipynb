```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizi = np.linspace(0,10,20)
```

```python
numpyDizi
```

```python
numpyDizi2 = numpyDizi ** 2
```

```python
numpyDizi2
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizi,numpyDizi2,"g",alpha=0.3)
benimEksen.plot(numpyDizi2,numpyDizi, color = "#123456")
# transparant(opacity) i√ßin alpha= 0.2
```

```python
(yeniFifur,yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizi, numpyDizi + 2, color= "#123456",linewidth = 3.0,linestyle=":")
yeniEksen.plot(numpyDizi,numpyDizi + 3, color= "yellow",linestyle="-.")
yeniEksen.plot(numpyDizi,numpyDizi + 4, color= "#013036",linewidth= 3.0,linestyle="--")
yeniEksen.plot(numpyDizi,numpyDizi + 5, color="#000000", marker="o",markerfacecolor="red",markersize=8)
```
