```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
KiloListesi = [20,30,40,50,60,65,70,90]
```

```python
kiloArray = np.array(KiloListesi)
```

```python
plt.plot(yasListesi,KiloListesi,"r")
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Kilo'nun yaşa göre oranı")
# diğer derleyicilerde plt.show() kullanılması gerekir. 
```

```python
numpyDizisi = np.linspace(0,10,11)
```

```python
numpyDizisi
```

```python
numpyDizisi2 = numpyDizisi ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi,numpyDizisi2,"g--")
```

```python
plt.plot(numpyDizisi,numpyDizisi2,"b*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi2,numpyDizisi,"r--")
plt.subplot(1,2,2) # 1 sıra olacak 2 kolon olacak, şu an 2. grafiği çiziyorum
plt.plot(numpyDizisi,numpyDizisi2,"r")
```

## Figure

```python
ilkFigure = plt.figure()
figureAxes = ilkFigure.add_axes([0.1,0.1,0.4,0.4])
figureAxes.plot(numpyDizisi,numpyDizisi2,"g")
figureAxes.set_xlabel("X")
figureAxes.set_ylabel("Y")
figureAxes.set_title("Baslik")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyDizisi,numpyDizisi2,"g")
eksen1.set_xlabel("X")
eksen1.set_ylabel("Y")
eksen1.set_title("Ana Grafik Baslık")

eksen2.plot(numpyDizisi2,numpyDizisi,"r--")
eksen2.set_xlabel("X")
eksen2.set_ylabel("Y")
eksen2.set_title("Küçük Grafik Baslık")
```

## subplots()

```python
(myFigure , myEksenler )= plt.subplots(nrows=1,ncols=2)
# myEksen.plot(numpyDizisi,numpyDizisi2,"g")
type(myEksenler)
for eksen in myEksenler:
    eksen.plot(numpyDizisi,numpyDizisi2,"g")
    eksen.set_xlabel("X ekseni")
# myFigure
plt.tight_layout()
```

```python
yeniFigür = plt.figure(dpi=75)# figsize=(6,5)
yeniEksen = yeniFigür.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi,numpyDizisi ** 2,label= "numpy dizisi ** 2")
yeniEksen.plot(numpyDizisi,numpyDizisi ** 3,label="numpy dizisi ** 3")
yeniEksen.legend(loc=2)
# yeniEksen.legend(loc=2) grafiğin sol üstündeki bilgilendirici ekranı gösterir.
# loc 1,2,3,4,5 değerleri alaiblir hazır
```
